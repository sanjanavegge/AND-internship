```python
good = card_data[card_data.fraud == 0]
bad = card_data[card_data.fraud == 1]
```

under sampling.....build samle dataset containing similar distibution of normal transaction and fradulant

```python
good_sample = good.sample(n=87403)
```

concate 2 data frame

```python
new_dataset = pd.concat([good_sample, bad], axis=0 )
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['fraud'].value_counts()
```

```python
new_dataset.groupby('fraud').mean()
```

spliting data into features ands targets(0,1)

```python
x = new_dataset.drop(columns='fraud',axis=1)
y=new_dataset['fraud']
print(x)
```

```python
print(y)
```

split data into training and testing

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

training

```python
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

```python
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction, y_train)
```

```python
 print('accuracy on training data: ', training_data_accuracy)
```

accuracy on test data

```python
x_test_prediction = model.predict(x_test)
test_data_accuracy =  accuracy_score(x_test_prediction, y_test)
```

```python
 print('accuracy on test data: ', test_data_accuracy)
```
